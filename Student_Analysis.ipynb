Student Performance Analysis

This project explores academic performance patterns among students and identifies factors influencing exam outcomes using exploratory data analysis and visualization techniques.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/StudentsPerformance.csv")

# Basic checks
print(df.info())
print(df.describe())
print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()

# Create total score column
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]

df.head()
```

```python
sns.histplot(df["total_score"], kde=True)
plt.title("Distribution of Total Student Scores")
plt.show()
```

```python
sns.boxplot(x="gender", y="total_score", data=df)
plt.title("Student Performance by Gender")
plt.show()
```

```python
sns.boxplot(x="parental level of education", y="total_score", data=df)
plt.xticks(rotation=45)
plt.title("Impact of Parental Education on Student Scores")
plt.show()
```

```python
sns.boxplot(x="lunch", y="total_score", data=df)
plt.title("Impact of Lunch Type on Performance")
plt.show()
```

```python
sns.boxplot(x="test preparation course", y="total_score", data=df)
plt.title("Impact of Test Preparation Course")
plt.show()
```

```python
sns.heatmap(df[["math score","reading score","writing score","total_score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Scores")
plt.show()
```
